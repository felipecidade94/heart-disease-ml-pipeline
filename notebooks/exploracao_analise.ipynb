### Análise das variáveis (Atributos)

##### Age (idade)

```python
df['Age'].value_counts()
```

```python
df['Age'].value_counts().sort_index()
```

```python
hist1 = px.histogram(df, x = 'Age',nbins=60)
hist1.update_layout(width=500, height=300, title_text='Distribuição por idade')
hist1.show()
```

```python
sns.histplot(df, x='Age', bins=30, color='orange', kde=True, stat='count')
```

##### Sex = Sexo

```python
df['Sex'].value_counts()
```

```python
sns.countplot(
   data=df,
   x='Sex',
   palette={
      'M': 'blue',
      'F': 'pink'
   }
)
plt.title('Distribuição por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Quantidade')
plt.show()
```

##### Chest Pain Type (Tipo de dor no peito)

ASY = assintomático  
NAP = dor não anginosa  
ATA = angina atípica  
TA = angina típica  
Angina = Dor no peito causada pela diminuição do fluxo de sangue no coração

```python
df['ChestPainType'].value_counts()
```

```python
sns.countplot(
   data=df,
   x='ChestPainType',
   palette='husl'
)
plt.title('Distribuição por Tipo de Dor no Peito')
plt.xlabel('Tipo de Dor no Peito')
plt.ylabel('Quantidade')
plt.show()
```

##### Resting BP (blood pressure) = pressão sanguínea em repouso em mmHg (sistólica -> maior)

```python
df.RestingBP.value_counts().sort_index()
```

```python
sns.histplot(df, x='RestingBP', bins=30, kde=True,stat='count', color='orange');
```

##### Cholesterol = colesterol sérico (mg/dl)

```python
df['Cholesterol'].value_counts().sort_index()
```

```python
sns.histplot(df, x= 'Cholesterol', bins=30, kde=True, stat='count', color='orange')
```

#### Fasting BS (blood sugar) = açucar no sangue em jejum (mg/dl)

0: FastingBS < 120 mg/dl (não diabético)  
1: FastingBS >= 120 mg/dl (diabético)

```python
df['FastingBS'].value_counts().sort_index()
```

```python
sns.countplot(
   data=df,
   x='FastingBS',
   palette={
      '1': 'red',
      '0': 'cyan'
   }
)
plt.title('Diabéticos x não diabéticos')
plt.xlabel('Estado')
plt.ylabel('Quantidade')
plt.show()
```

##### Resting ECG = eletrocardiograma em repouso

Normal = normal
LVH = hipertrofia ventricular esquerda
ST = Anormalidade de onda ST-T

```python
df['RestingECG'].value_counts().sort_index()
```

```python
sns.countplot(data=df,
   x='RestingECG',
   palette='husl'
)
plt.title('Distribuição eletrocardiograma em repouso')
plt.xlabel('Estado')
plt.ylabel('Quantidade')
plt.show();
```
